### Shorten author names correctly 

```python
with open(bibname+".bib", "r") as f:
    bib = f.readlines()
```

```python
full_name_pattern = r"([A-Z]\S*, [A-Z]\S* [A-Z]\S*)|([A-Z]\S*, [A-Z]\S*)"
polished_bib = []
for line in bib: 
    match_list = list(re.finditer(full_name_pattern, line[:-3])) #  -3 removes problems at the end of 
                                                                 #  the string with '}
    replacements = []
    for match in match_list:
        if match.group(1) is not None:
            # we have a full tripartite name
            name = match.group(1)
            name_parts = name.split()
            edited_name = name_parts[0]+" "
            edited_name += name_parts[1][0]+". " + name_parts[2][0]+"."
            print(name, ">", edited_name)
            replacements.append((name, edited_name))
        if match.group(2) is not None:
            # we have a shortened name
            name = match.group(2)
            name_parts = name.split()
            edited_name = name_parts[0]+" "
            if re.match(r"[A-Z]{2}", name_parts[1]):
                # we have a bipartite name-middle name
                edited_name += name_parts[1][0]+". " + name_parts[1][1]+"."
            else:
                edited_name += name_parts[1][0]+". "
            print(name, ">", edited_name)
            replacements.append((name, edited_name))

    for replacement in replacements:
        line = line.replace(replacement[0], replacement[1])
    polished_bib.append(line)
```

```python
with open(f"{bibname}-polished.bib", "w") as f:
    f.writelines(polished_bib)
```
